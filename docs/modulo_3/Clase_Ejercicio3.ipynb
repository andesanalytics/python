```python
#Cargar BBDD
df = pd.read_excel('/content/Seguros_Auto.xlsx', sheet_name = 'Datos Extras')
```

```python
# Head ayuda a una visualizaci√≥n r√°pida
df.tail()
# Info me ayuda a confirmar el tipo de objeto de las columnas
df.info()
```

```python
# La fecha al usar strftime se convierte en un objeto de tipo str
df['FECHA_PAGARE'] = df['FECHA_PAGARE'].dt.strftime('%Y-%m-%d')
# Para convertir un str a fecha se usa to_datetime()
df['FECHA_PAGARE'] = pd.to_datetime(df['FECHA_PAGARE'])
```

```python
# Calcular el plazo total del seguro, es decir, desde que fue contratado, hasta que termino
df['Dif'] = df['Fecha de Termino'] - df['FECHA_PAGARE']
df.head()
df.info()
```

```python
#Queremos crear una columna que se llame Dif meses y que convierta la Diferencia en dias en meses
df['DIF Meses'] = df['Dif']/30.44
```

```python
df_Multi = pd.read_excel('/content/Seguros_Auto.xlsx', sheet_name= 'Plazo MultiAsistencia')
df_Multi.head()
df_Multi.info()
```

```python
# Unir dos BBDD
df_full = pd.merge(df,
                   df_Multi,
                   on = 'ID M√≥naco',
                   how = 'right')

df_full = df_full[['ID M√≥naco', 'FECHA_PAGARE','Fecha de Termino', 'Plazo_x', 'Periodo', 'Monto Asistencia']]
df_full.rename(columns = {'Plazo_x': 'Plazo'}, inplace = True)
df_full.head()
```
